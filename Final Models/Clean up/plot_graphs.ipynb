```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
with open('SSPSO.csv', 'r') as read_obj: # read csv file as a list of lists
    csv_reader = csv.reader(read_obj) # pass the file object to reader() to get the reader object
    SSPSO = list(csv_reader) # Pass reader object to list() to get a list of lists
# print(SSPSO[4])
SSPSOVals = [eval(i) for i in SSPSO[4]]

with open('SLPSO.csv', 'r') as read_obj: # read csv file as a list of lists
    csv_reader = csv.reader(read_obj) # pass the file object to reader() to get the reader object
    SLPSO = list(csv_reader) # Pass reader object to list() to get a list of lists
# print(SLPSO[4])
SLPSOVals = [eval(i) for i in SLPSO[4]]

with open('SLPSO1.csv', 'r') as read_obj: # read csv file as a list of lists
    csv_reader = csv.reader(read_obj) # pass the file object to reader() to get the reader object
    SLPSO1 = list(csv_reader) # Pass reader object to list() to get a list of lists
```

```python
graphEpochCounter = range(100)
fig , (ax) = plt.subplots(ncols=1, nrows=1)
# ax.fill_between(graphEpochCounter, min_fit, max_fit, alpha=0.2)
ax.plot(graphEpochCounter, SSPSOVals, label="SSPSO")
ax.plot(graphEpochCounter, SLPSOVals, label="SLPSO", linestyle="dashed")
# ax.plot(graphEpochCounter, [eval(i) for i in SLPSO1[4]], label="SLPSO", linestyle="dotted")
ax.set_title("Max Training Accuracy")
ax.set_xlabel('Generations')
ax.set_ylabel("Accuracy")
ax.legend()
figure = ax.get_figure()
plt.show()
figure.savefig("SLSPSO-accuracy.pdf", bbox_inches='tight')
figure.savefig("SLSPSO-accuracy.png", dpi=300, bbox_inches='tight')
```

```python
SSPSOVals1 = [eval(i) for i in SSPSO[9]]
SLPSOVals1 = [eval(i) for i in SLPSO[9]]

graphEpochCounter = range(100)
fig , (ax) = plt.subplots(ncols=1, nrows=1)
# ax.fill_between(graphEpochCounter, min_fit, max_fit, alpha=0.2)
ax.plot(graphEpochCounter, SSPSOVals1, label="SSPSO")
ax.plot(graphEpochCounter, SLPSOVals1, label="SLPSO", linestyle="dashed")
# ax.plot(graphEpochCounter, [eval(i) for i in SLPSO1[4]], label="SLPSO", linestyle="dotted")
ax.set_title("Min Training Loss")
ax.set_xlabel('Generations')
ax.set_ylabel("Loss")
ax.legend()
figure = ax.get_figure()
plt.show()
figure.savefig("SLSPSO-loss.pdf", bbox_inches='tight')
figure.savefig("SLSPSO-loss.png", dpi=300, bbox_inches='tight')
```

```python
graphEpochCounter = range(100)
fig , (ax) = plt.subplots(ncols=1, nrows=1)
ax.fill_between(graphEpochCounter, [eval(i) for i in SSPSO[3]], [eval(i) for i in SSPSO[4]], alpha=0.2)
ax.fill_between(graphEpochCounter, [eval(i) for i in SLPSO[3]], [eval(i) for i in SLPSO[4]], alpha=0.2)
ax.plot(graphEpochCounter, [eval(i) for i in SSPSO[2]], label="SSPSO")
ax.plot(graphEpochCounter, [eval(i) for i in SLPSO[2]], label="SLPSO", linestyle="dashed")
ax.set_title("Average Training Accuracy")
ax.set_xlabel('Generations')
ax.set_ylabel("Accuracy")
ax.legend()
figure = ax.get_figure()
plt.show()
figure.savefig("SLSPSO-avg-accuracy.pdf", bbox_inches='tight')
figure.savefig("SLSPSO-avg-accuracy.png", dpi=300, bbox_inches='tight')
```
